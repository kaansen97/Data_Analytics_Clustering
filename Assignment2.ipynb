# Cluster modeling

## k-means

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
# Encode the 'From' column
label_encoder = LabelEncoder()
data['From_Encoded'] = label_encoder.fit_transform(data['From'])

# Print the encoded values and their corresponding original values
encoded_values = data['From_Encoded'].unique()
original_values = label_encoder.inverse_transform(encoded_values)

for encoded, original in zip(encoded_values, original_values):
    print(f"Encoded: {encoded} -> Original: {original}")
```

```python
# Construct the feature matrix
X = data[["Article_ID","Word_ID", "Occurence_Number","From_Encoded"]]
X_new = resample(X, n_samples=int(0.2 * len(X)), random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X_new[["Article_ID","Word_ID", "Occurence_Number"]],X_new[["From_Encoded"]], test_size=0.2, random_state=42)

# Print the shapes of the train and test sets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Finding the optimal number of clusters
max_clusters = 10  # Maximum number of clusters
best_cluster_count = 0
best_silhouette_score = -1
inertia = []
score=[]
for k in range(2, max_clusters + 1):
    # Construction of K-means model
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X_train)

    # Calculation of contour coefficient
    silhouette_avg = silhouette_score(X_train, kmeans.labels_)
    score.append(silhouette_avg)
        # Calculate the total internal sum of squares
    inertia.append(kmeans.inertia_)
    # Select the number of clusters with the highest profile coefficients
    if silhouette_avg > best_silhouette_score:
        best_silhouette_score = silhouette_avg
        best_cluster_count = k
```

```python
import matplotlib.pyplot as plt# Drawing the elbow
plt.plot(range(1, max_clusters ), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Total Within-Cluster Sum of Squares')
plt.title('Elbow Method')
plt.show()
```

```python
# Set the number of clusters to 4
k = 4

# Construct the K-means model
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=200, n_init=20, random_state=0)
kmeans.fit(X_train)

# Compute silhouette score
silhouette_avg = silhouette_score(X_train, kmeans.labels_)

# Compute total within-cluster sum of squares
inertia = kmeans.inertia_

# Update best silhouette score and cluster count
if silhouette_avg > best_silhouette_score:
    best_silhouette_score = silhouette_avg
    best_cluster_count = k

# Plot the distribution of cluster centers and sample points
plt.scatter(X_train.iloc[:, 0], X_train.iloc[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red')
plt.xlabel('Article_ID')
plt.ylabel('Word_ID')
plt.title('Distribution of Cluster Centers and Sample Points')
plt.show()
```

## Agglomerative Cluster

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
# Find the optimal number of clusters
max_clusters = 5  # Maximum number of clusters
best_cluster_count = 0
best_silhouette_score = -1
silhouette_scores = []
for k in range(2, max_clusters + 1):
    # Constructing hierarchical clustering models (cohesive clustering)
    hierarchical = AgglomerativeClustering(n_clusters=k)
    hierarchical.fit(X_train)

    # Calculation of contour coefficient
    silhouette_avg = silhouette_score(X_train, hierarchical.labels_)
    silhouette_scores.append(silhouette_avg)
    # Select the number of clusters with the highest profile coefficients
    if silhouette_avg > best_silhouette_score:
        best_silhouette_score = silhouette_avg
        best_cluster_count = k

# Output the optimal number of clusters
print("Best number of clusters:", best_cluster_count)
```

```python
plt.plot(range(2, max_clusters + 1), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.title('Silhouette Score for Different Cluster Numbers')
plt.show()
```

```python
# Building hierarchical clustering Model
hierarchical = AgglomerativeClustering(n_clusters=4)
hierarchical.fit(X_train)
labels = hierarchical.labels_
plt.scatter(X_train["Article_ID"], X_train["Word_ID"], c=labels)
plt.scatter(hierarchical.children_[:, 0], hierarchical.children_[:, 1], c='r', marker='o', label='Cluster centers')
plt.xlabel("Article_ID")
plt.ylabel("Word_ID")
plt.title("Clustering Results")
plt.legend()
plt.show()
```

## GMM Cluster

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score

max_clusters = 5  # Maximum number of clusters
best_cluster_count = 0
best_silhouette_score = -1
silhouette_scores2 = []

for k in range(2, max_clusters + 1):
    # Building the GMM model
    gmm = GaussianMixture(n_components=k)
    gmm.fit(X_train)

    # Calculation of contour coefficient
    silhouette_avg = silhouette_score(X_train, gmm.predict(X_train))
    silhouette_scores2.append(silhouette_avg)

    # Select the number of clusters with the highest profile coefficients
    if silhouette_avg > best_silhouette_score:
        best_silhouette_score = silhouette_avg
        best_cluster_count = k

# Output the optimal number of clusters
print("Best number of clusters:", best_cluster_count)
```

```python
plt.plot(range(2, max_clusters + 1), silhouette_scores2, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.title('Silhouette Score for Different Cluster Numbers')
plt.show()
```

```python

# Construct the GMM model
gmm = GaussianMixture(n_components=2)
gmm.fit(X_train)

# Compute silhouette score
silhouette_avg = silhouette_score(X_train, gmm.predict(X_train))

# Update best silhouette score and cluster count
best_silhouette_score = silhouette_avg
best_cluster_count = k

# Plot the distribution of cluster centers and sample points
plt.scatter(X_train.iloc[:, 0], X_train.iloc[:, 1], c=gmm.predict(X_train), cmap='viridis')
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], marker='x', color='red')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Distribution of Cluster Centers and Sample Points')
plt.show()

# Output the best number of clusters
print("Best number of clusters:", best_cluster_count)
```
